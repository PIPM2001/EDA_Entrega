1. Importamos el histórico del Euribor medio mensual. 

   Fuente: Banco de España

```python
df_1 = pd.read_excel("../data/be1901EXCEL.xlsx")
fuente_1 = "Banco de España"
```

Nos quedamos con la columnas que nos interesan, "ALIAS DE LA SERIE" que es el periodo y  la columna "BE_19_1.5" que es el Euribor a 1 año.

```python
df_1 = df_1.loc[:,["ALIAS DE LA SERIE", "BE_19_1.5"]]
```

Cambiamos el nombre de las columnas por uno mas apropiado.

```python
df_1 = df_1.rename(columns={"ALIAS DE LA SERIE": "Periodo",  "BE_19_1.5": "Euribor_1y"})
```

```python
df_1.info()
```

```python
df_1
```

La función info nos informa que no hay nulos en "Euribor_1y", pero lo cierto es que si los hay. Lo que ocurre es que están informados como "-".
Eliminamos las filas sin dato de Euribor_1y

```python
df_1[df_1["Euribor_1y"] == "-"]
```

¿Por qué vemos estos nulos? Por que en realidad el Euribor no existió hasta enero de 1999. En esa fecha arrancó el sistema europeo de credito interbancario. Previamente cada país tenía su propio sistema al igual que su propia moneda.

```python
df_1.drop(df_1[df_1["Euribor_1y"] == "-"].index, inplace=True)
```

```python
df_1.info()
```

Cambiamos el tipo de dato de "Euribor_1y" a float.

```python
df_1['Euribor_1y'] = df_1['Euribor_1y'].astype('float64')
df_1.info()
```

Comprobamos que no hay duplicados

```python
df_1.duplicated().value_counts()
```

Cada fuente de datos utilza una nomemclatura para el periodo:

Banco de españa = "ENE 2024"

BCE = "Jan 2024" 

INE = "2024M01" 

Esto será un problema cuando vayamos a fusionar los datos en un unico dataframe ya que no habrá una clave común por la que asociarlos. 

**Vamos a crear la clave común que tendrá el formato AAAAMM**

```python
meses = {"ENE" : "01",
         "FEB" : "02",
         "MAR" : "03",
         "ABR" : "04",
         "MAY" : "05",
         "JUN" : "06",
         "JUL" : "07",
         "AGO" : "08",
         "SEP" : "09",
         "OCT" : "10",
         "NOV" : "11",
         "DIC" : "12",
         }
def Idmes(abr_mes):
    return meses[abr_mes]
df_1["IdPeriodo"]=df_1.Periodo.str[-4:] + df_1.Periodo.str[0:3].apply(Idmes)

df_1
```

```python
df_1.info()
```

2. Importamos el histórico del tipo oficial del dinero diario.

   Fuente: Banco Central Europeo

```python
df_2 =pd.read_csv("../data/ECB Data Portal_20240405225205.csv")
fuente_2 = "Banco Central europeo"
```

```python
df_2.info()
```

```python
df_2.head()
```

```python
df_2.rename(columns={"Deposit facility - date of changes (raw data) - Level (FM.D.U2.EUR.4F.KR.DFR.LEV)": "Int_rate"}, inplace=True)
```

```python
df_2.info()
```

No apreciamos que haya valores nulos.

El tipo de interes es un float, lo cual es correcto.

Vamos a comprobar que no hay duplicados

```python
df_2.duplicated().value_counts()
```

Vamos a crear la clave común que tendrá el formato AAAAMM

```python
df_2["IdPeriodo"]=df_2.DATE.str[0:4] + df_2.DATE.str[5:7]
df_2
```

Del Euribor no tenemos datos diarios sino las medias mensuales, lo cual es apropiado para nuestro análisis de su influencia en el precio de las hipotecas ya que la renovación del tipo cada año se calcula con la Euribor medio mensual. Pero del tipo oficial del BCE tenemos el datos diario. Para poder tener dos variables comparables vamos a quedarnos con las medias mensuales.

```python
df_2 = df_2.groupby("IdPeriodo").mean("Int_rate")
df_2
```
